```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read csv file

netflix_df = pd.read_csv("netflix_data.csv")
netflix_df
```

```python
# filter dataset to only show netflix movies

netflix_subset = netflix_df[netflix_df["type"] == "Movie"]
netflix_subset
```

```python
# filter dataset to only show certain columns

netflix_movies = netflix_subset[["title", "country", "genre", "release_year", "duration"]]
netflix_movies
```

```python
# new dataframe for movies that are less than 60 minutes

short_movies = netflix_movies[netflix_movies["duration"] < 60]
short_movies
```

```python
# adding color for different genre of movies
colors = [] 

for label, row in netflix_movies.iterrows():
    if row["genre"] == "Children":
        colors.append("Green")
    elif row["genre"] == "Documentaries":
        colors.append("Black")
    elif row["genre"] == "Stand_Up":
        colors.append("Blue")
    else:
        colors.append("Blue")

colors[:10]
```

```python
# initialize new figure
fig = plt.figure(figsize=(12,8))

# create scatter plot
netflix_movies.plot(kind = 'scatter', x = 'release_year', y = 'duration', c = colors)

# Create title and labels
plt.title("Movie Duration by Year of Release")
plt.xlabel("Release Year")
plt.ylabel("Duration (min)")

# Show the plot
plt.show()
```

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# read csv file
netflix_df = pd.read_csv("netflix_data.csv")
netflix_df

# filter dataset to only show netflix movies
netflix_subset = netflix_df[netflix_df["type"] == "Movie"]
netflix_subset

# filter dataset to only show certain columns
netflix_movies = netflix_subset[["title", "country", "genre", "release_year", "duration"]]
netflix_movies

# new dataframe for movies that are less than 60 minutes
short_movies = netflix_movies[netflix_movies["duration"] < 60]
short_movies

# adding color for different genre of movies
colors = [] 

for label, row in netflix_movies.iterrows():
    if row["genre"] == "Children":
        colors.append("Green")
    elif row["genre"] == "Documentaries":
        colors.append("Black")
    elif row["genre"] == "Stand_Up":
        colors.append("Blue")
    else:
        colors.append("Blue")

colors[:10]

# initialize new figure
fig = plt.figure(figsize=(12,8))
       
# create scatter plot
netflix_movies.plot(kind='scatter', x='release_year', y='duration', c=colors)

# Create title and labels
plt.title = "Movie Duration by Year of Release" 
plt.xlabel = "Release Year"  
plt.ylabel = "Duration (min)"  

# Show the plot
plt.show()

# answer to question
answer = "maybe"
```
